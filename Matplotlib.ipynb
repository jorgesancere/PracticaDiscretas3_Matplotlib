## 7.2. Creating a Plot

Ahora veremos como crear una línea usando matplotlib. Con la funcion .plot, podemos introducir dos datos ( en este caso listas), como puede ser una lista para el eje x y otra para el eje y sea crea una linea que pasa por los datos introducidos.
Con la funcion .show, mostramos la línea

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]

plt.figure(figsize=(10, 5)) # Create a figure with a size of 10x5, width x height
plt.plot(x, y) # Create a line plot
plt.show() # Show the plot
```

También podemos crear gráficas en tres dimensiones, usando la funcion .plot, que ya hemos visto anteriormente

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
# This part is necessary to create the 3D plot
fig = plt.figure(figsize=(10, 5)) # Create a figure with a size of 10x5, width x height
ax = fig.add_subplot(111, projection='3d') # Create a 3D axis, 111 means 1x1 grid, first subplot, projection='3d' means 3D plot

x = np.random.standard_normal(100) # Create random data
y = np.random.standard_normal(100) # Create random data
z = np.random.standard_normal(100) # Create random data

ax.scatter(x, y, z) # Create a 3D scatter plot
plt.show() # Show the plot
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(11)
# This part is necessary to create the 3D plot
fig = plt.figure(figsize=(10, 20)) # Create a figure with a size of 10x5, width x height
ax = fig.add_subplot(111, projection='3d') # Create a 3D axis, 111 means 1x1 grid, first subplot, projection='3d' means 3D plot

x = np.random.standard_normal(100) # Create random data
y = np.random.standard_normal(100) # Create random data
z = np.random.standard_normal(100) # Create random data

ax.plot(x, y, z) 
plt.show() # Show the plot
```

Además de los anteriores, matplotlib también acepta graficos de barras, gracias a la funcion .bar:

Finalmente, podemos crear histogramas, con la funcion .hist, con el que podemos ver la frecuencia de los datos introducidos: 

```python

import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
data = np.random.randn(1000) # Create random data

plt.hist(data, bins=30) # Create a histogram with 30 bins. Here bins are the number of intervals in the data, and it is a parameter of the histogram
                        # if bins is too small, the histogram will be too smooth, and if bins is too large, the histogram will be too spiky

plt.show() # Show the plot
```

```python

import matplotlib.pyplot as plt
import numpy as np
np.random.seed(45)
data = np.random.randn(100)

plt.hist(data, bins=5)
plt.show() # Show the plot
```

## 7.3. Customizing the Plot

Ahora veremos cómo podemos personalizar nuestras gráficas de matplotlib:

### 7.3.1. Adding a Title

Gracias a la ufncion .title, podemos hacer que nuestra gráfica tenga título.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
plt.figure(figsize=(10, 5)) # Create a figure with a size of 10x5, width x height
plt.plot(x, y) # Create a line plot
plt.title('Simple Line Plot') # Add a title to the plot
plt.show() # Show the plot
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 4, 3, 4, 3, 2]
plt.figure(figsize=(10, 5)) # Create a figure with a size of 10x5, width x height
plt.plot(x, y) # Create a line plot
plt.scatter(x, y)
plt.title('Mi gráfica')
plt.show() 
```

### 7.3.2. Adding Labels

También podemos añadir etiquetas a cada eje, gracias a las funciones xlabel y ylabel:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
plt.figure(figsize=(10, 5)) # Create a figure with a size of 10x5, width x height
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 4, 3, 4, 3, 2]
plt.figure(figsize=(10, 5)) # Create a figure with a size of 10x5, width x height
plt.plot(x, y) # Create a line plot
plt.scatter(x, y)
plt.title('Mi gráfica')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show() 
```

### 7.3.3. Adding a Legend

Gracias a la función .legend, podemos crear leyendas, en las que aparezcan que es cada cosa de nuestra gráfica:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
plt.figure(figsize=(10, 5)) # Create a figure with a size of 10x5, width x height
plt.plot(x, y, label='Line 1')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 4, 3, 4, 3, 2]
plt.figure(figsize=(10, 5)) # Create a figure with a size of 10x5, width x height
plt.plot(x, y, label = 'Línea') # Create a line plot
plt.scatter(x, y, label = 'Puntos')
plt.title('Mi gráfica')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show() 
```

### 7.3.4. Changing the Color, Line Style, and Marker Style

Ahora veremos como cambiar tanto el color, como el estilo de las lineas y el de los puntos en las coordenadas: 

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]

plt.figure(figsize=(10, 5)) # Create a figure with a size of 10x5, width x height
plt.plot(x, y, color='red', linestyle='--', marker='o')

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 4, 3, 4, 3, 2]
plt.figure(figsize=(10, 5)) # Create a figure with a size of 10x5, width x height
plt.plot(x, y, label = 'LíneaM', color='violet', linestyle=(0, (5, 1)), marker = 'D') # Create a line plot
plt.title('Mi gráfica')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show() 
```

## 7.4. Multiple Plots

### 7.4.1. Raw version

Gracias a la función .subplot, podemos crear varios plots en una misma figura:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
plt.figure(figsize=(10, 5)) # Create a figure with a size of 10x5, width x height
plt.subplot(1, 2, 1)
plt.plot(x, y)

plt.subplot(1, 2, 2)
plt.scatter(x, y)

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 4, 3, 4, 3, 2]
plt.figure(figsize=(6, 6)) # Create a figure with a size of 10x5, width x height
plt.subplot(2, 1, 1)
plt.plot(x, y, label = 'LíneaM', color='violet', linestyle=(0, (5, 1)), marker = 'D') # Create a line plot
plt.legend()

plt.subplot(2, 1, 2)
plt.scatter(x, y, label = 'PuntosM', color='Blue', marker = '+') # Create a line plot
plt.legend()

plt.show()
```

### 7.4.2. Formatted version

Gracias a las funciones ax y fig, podemos crear varias figuras y posteriormente, 

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]

fig, ax = plt.subplots(1, 2, figsize=(10, 5)) # Create a figure and 2 axes. Note that figsize is the size of the figure
ax[0].plot(x, y, label='Line 1') # Plot in the first axis, and add a label
ax[0].set_title('Line Plot') # We can also add a title to the plot
# We can also add labels to the x-axis and y-axis
ax[0].set_xlabel('X-axis')
ax[0].set_ylabel('Y-axis')
# We can also change the color, the line style, and the marker style of the plot
ax[0].plot(y, x, color='red', linestyle='dashdot', marker='x', label='Line 2')
# We can also add a legend to the plot
ax[0].legend(['Line 1', 'Line 2'])

ax[1].scatter(x, y, label='Scatter 1') # Plot in the second axis
ax[1].set_title('Scatter Plot')
ax[1].set_xlabel('X-axis')
ax[1].set_ylabel('Y-axis')
ax[1].scatter(y, x, color='green', marker='o', label='Scatter 2')
ax[1].legend(['Scatter 1', 'Scatter 2'])
plt.show()
```
